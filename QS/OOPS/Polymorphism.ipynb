Notes :- 
Polymorphism - It is a phenomenon of making a single operator or a method to work on multiple operations. 

We can achieve the polymorphism using the concept like method overloading, operator overloading

Method Overloading - It is the phenomenon of making a single method to work on multiple operations .
In python we cannot perform method overloading , if we try to perform method overloading by default it will perform method overriding.

Method Overriding - If we have 2 method in a same class with a same name the old method will be overwritten with the latest method .

Operator Overloading - It is a phenomenon of making a single operator to work on multiple operations .
We can make any operator to support for the objects which are created for user defined class with the help of following dundered/magic methods


# Python Operators and Their Magic Methods

## I. Arithmetic Operators

| Operator       | Operation         | Magic Method     | Symbol |
|----------------|-------------------|------------------|--------|
| Addition       | `obj1 + obj2`     | `__add__`        | `+`    |
| Subtraction    | `obj1 - obj2`     | `__sub__`        | `-`    |
| Multiplication | `obj1 * obj2`     | `__mul__`        | `*`    |
| True Division  | `obj1 / obj2`     | `__truediv__`    | `/`    |
| Floor Division | `obj1 // obj2`    | `__floordiv__`   | `//`   |
| Modulus        | `obj1 % obj2`     | `__mod__`        | `%`    |
| Power          | `obj1 ** obj2`    | `__pow__`        | `**`   |

## II. Bitwise Operators

| Operator       | Operation         | Magic Method     | Symbol |
|----------------|-------------------|------------------|--------|
| Bitwise AND    | `obj1 & obj2`     | `__and__`        | `&`    |
| Bitwise OR     | `obj1 \| obj2`    | `__or__`         | `\|`    |
| Bitwise XOR    | `obj1 ^ obj2`     | `__xor__`        | `^`    |
| Bitwise NOT    | `~obj1`           | `__invert__`     | `~`    |
| Left Shift     | `obj1 << obj2`    | `__lshift__`     | `<<`   |
| Right Shift    | `obj1 >> obj2`    | `__rshift__`     | `>>`   |


## III. Relational (Comparison) Operators

| Operator                   | Operation           | Magic Method     | Symbol |
|----------------------------|---------------------|------------------|--------|
| Equal To                   | `obj1 == obj2`      | `__eq__`         | `==`   |
| Not Equal To               | `obj1 != obj2`      | `__ne__`         | `!=`   |
| Greater Than               | `obj1 > obj2`       | `__gt__`         | `>`    |
| Less Than                  | `obj1 < obj2`       | `__lt__`         | `<`    |
| Greater Than or Equal To   | `obj1 >= obj2`      | `__ge__`         | `>=`   |
| Less Than or Equal To      | `obj1 <= obj2`      | `__le__`         | `<=`   |


```python
class Demo:
    def __init__(self,num):
        self.num = num
        
x = Demo(10)
y = Demo(20)
print(x+y)
```

```python
class Demo:
    def __init__(self,num):
        self.num = num
        
    def __add__(self,other):
        return self.num + other.num
    
    def __sub__(self,other):
        return self.num - other.num
    
    def __mul__(self,other):
        return self.num * other.num
    
    def __truediv__(self,other):
        return self.num / other.num
    
    def __floordiv__(self,other):
        return self.num // other.num
    
    def __mod__(self,other):
        return self.num % other.num
    
    def __pow__(self,other):
        return self.num ** other.num
    
    def __and__(self,other):
        return self.num % other.num
    
    def __or__(self,other):
        return self.num | other.num
    
    def __xor__(self,other):
        return self.num ^ other.num
    
    def __invert__(self):
        return ~self.num
    
    def __lshift__(self,other):
        return self.num << other.num
    
    def __rshift__(self,other):
        return self.num >> other.num
    
    def __eq__(self,other):
        return self.num == other.num
    
    def __ne__(self,other):
        return self.num != other.num
    
    def __gt__(self,other):
        return self.num > other.num
    
    def __lt__(self,other):
        return self.num < other.num
    
    def __ge__(self,other):
        return self.num >= other.num
    
    def __le__(self,other):
        return self.num <= other.num
    
    
x = Demo(100)
y = Demo(2)
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x//y)
print(x%y)
print(x**y)
print(x&y)
print(x|y)
print(x^y)
print(~x)
print(x<<y)
print(x>>y)
print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)

    
    
```
