## String Manipulation Functions in Python

Python offers a powerful set of built-in functions designed for efficient manipulation of strings. These functions streamline common string operations, making your code cleaner and more effective.

-----

### Case Conversion Functions

These functions are used to alter the case of characters within a string.

  * **`upper()`**:

      * **Purpose**: Converts all lowercase alphabets in the string to **uppercase**. Other characters (numbers, symbols, spaces) remain unchanged.
      * **Returns**: A new string with all alphabetic characters converted to uppercase.
      * **Example**:
        ```python
        my_string = "Hello World123!"
        print(my_string.upper())
        # Output: HELLO WORLD123!
        ```

  * **`lower()`**:

      * **Purpose**: Converts all uppercase alphabets in the string to **lowercase**. Other characters remain unchanged.
      * **Returns**: A new string with all alphabetic characters converted to lowercase.
      * **Example**:
        ```python
        my_string = "HELLO World123!"
        print(my_string.lower())
        # Output: hello world123!
        ```

  * **`capitalize()`**:

      * **Purpose**: Converts the **first character of the entire string to uppercase** and all remaining alphabetic characters to lowercase.
      * **Returns**: A new string with the first character capitalized.
      * **Example**:
        ```python
        my_string = "hello world"
        print(my_string.capitalize())
        # Output: Hello world

        my_string = "tHIS iS a TeST"
        print(my_string.capitalize())
        # Output: This is a test
        ```

  * **`title()`**:

      * **Purpose**: Converts the **first character of each word** in the string to uppercase and all remaining alphabetic characters in that word to lowercase. Words are typically separated by spaces or other non-alphanumeric characters.
      * **Returns**: A new string in title case format.
      * **Example**:
        ```python
        my_string = "this is a great example"
        print(my_string.title())
        # Output: This Is A Great Example

        my_string = "hello-world_123"
        print(my_string.title())
        # Output: Hello-World_123
        ```

  * **`swapcase()`**:

      * **Purpose**: **Swaps the case** of all alphabetic characters in the string. Uppercase letters become lowercase, and lowercase letters become uppercase. Other characters remain unchanged.
      * **Returns**: A new string with the cases swapped.
      * **Example**:
        ```python
        my_string = "HeLlO wOrLd123!"
        print(my_string.swapcase())
        # Output: hElLo WoRlD123!
        ```

### String Content Check Functions (Boolean Return)

These functions return a boolean value (`True` or `False`) indicating whether the string meets a specific condition.

  * **`isupper()`**:

      * **Purpose**: Checks if all **alphabetic characters** in the string are uppercase. Non-alphabetic characters are ignored.
      * **Returns**: `True` if all alphabetic characters are uppercase; `False` otherwise (e.g., if any lowercase alphabet is present).
      * **Example**:
        ```python
        print("HELLO".isupper())
        # Output: True
        print("hello".isupper())
        # Output: False
        print("HELLO123".isupper())
        # Output: True (numbers are ignored)
        ```

  * **`islower()`**:

      * **Purpose**: Checks if all **alphabetic characters** in the string are lowercase. Non-alphabetic characters are ignored.
      * **Returns**: `True` if all alphabetic characters are lowercase; `False` otherwise.
      * **Example**:
        ```python
        print("hello".islower())
        # Output: True
        print("HELLO".islower())
        # Output: False
        print("hello123".islower())
        # Output: True
        ```

  * **`isnumeric()`**:

      * **Purpose**: Checks if all characters in the string are **numeric** (digits 0-9).
      * **Returns**: `True` if the string contains only numeric characters; `False` otherwise. This includes digits, fractions, subscripts, superscripts, etc.
      * **Example**:
        ```python
        print("12345".isnumeric())
        # Output: True
        print("12.34".isnumeric())
        # Output: False (contains a dot)
        print("Â½".isnumeric())
        # Output: True (unicode numeric character)
        ```

  * **`isalpha()`**:

      * **Purpose**: Checks if all characters in the string are **alphabetic** (a-z, A-Z). Spaces, numbers, and symbols are not considered alphabetic.
      * **Returns**: `True` if the string contains only alphabetical characters; `False` otherwise.
      * **Example**:
        ```python
        print("HelloWorld".isalpha())
        # Output: True
        print("Hello World".isalpha())
        # Output: False (contains space)
        ```

  * **`isalnum()`**:

      * **Purpose**: Checks if all characters in the string are **alphanumeric** (alphabetic or numeric). This includes letters (a-z, A-Z) and digits (0-9). Spaces and symbols are not alphanumeric.
      * **Returns**: `True` if the string contains only alphanumeric characters; `False` otherwise.
      * **Example**:
        ```python
        print("HelloWorld123".isalnum())
        # Output: True
        print("Hello World".isalnum())
        # Output: False (contains space)
        ```

  * **`isidentifier()`**:

      * **Purpose**: Checks if the string is a **valid Python identifier**. A valid identifier must start with a letter (a-z, A-Z) or an underscore (`_`), and subsequent characters can be letters, numbers, or underscores. It cannot be a Python keyword.
      * **Returns**: `True` if the string is a valid identifier; `False` otherwise.
      * **Example**:
        ```python
        print("my_variable".isidentifier())
        # Output: True
        print("123variable".isidentifier())
        # Output: False (starts with a number)
        print("class".isidentifier())
        # Output: False (keyword)
        ```

  * **`istitle()`**:

      * **Purpose**: Checks if the given string is in **title case format**. This means that the first letter of each word is uppercase, and all other letters are lowercase. Words are separated by non-alphabetic characters.
      * **Returns**: `True` if the string is in title format; `False` otherwise.
      * **Example**:
        ```python
        print("Hello World".istitle())
        # Output: True
        print("Hello world".istitle())
        # Output: False (second word not capitalized)
        ```

### Search and Replace Functions

These functions help in finding and manipulating substrings within a string.

  * **`count()`**:

      * **Purpose**: Counts the number of times a specified **substring** is repeated within a given string, optionally within a specified start and end index.
      * **Syntax**: `string.count(substring, [start_index], [end_index])`
          * `substring`: The string to search for.
          * `start_index` (optional): The starting position of the search. Defaults to `0`.
          * `end_index` (optional): The ending position of the search (exclusive). Defaults to the end of the string.
      * **Returns**: An integer representing the number of times the `substring` appears.
      * **Example**:
        ```python
        my_string = "banana_apple_banana_orange"
        print(my_string.count("banana"))
        # Output: 2
        print(my_string.count("a", 7, 20))
        # Output: 3 (counts 'a' from index 7 to 19)
        ```

  * **`startswith()`**:

      * **Purpose**: Checks if the string **starts with** the specified substring.
      * **Syntax**: `string.startswith(prefix, [start_index], [end_index])`
          * `prefix`: The substring to check for at the beginning of the string.
          * `start_index` (optional): The index from which to start the check.
          * `end_index` (optional): The index up to which to end the check (exclusive).
      * **Returns**: `True` if the string starts with the `prefix`; `False` otherwise.
      * **Example**:
        ```python
        my_string = "Hello World"
        print(my_string.startswith("Hello"))
        # Output: True
        print(my_string.startswith("World", 6))
        # Output: True (checks from index 6)
        ```

  * **`endswith()`**:

      * **Purpose**: Checks if the string **ends with** the specified substring.
      * **Syntax**: `string.endswith(suffix, [start_index], [end_index])`
          * `suffix`: The substring to check for at the end of the string.
          * `start_index` (optional): The index from which to start the check.
          * `end_index` (optional): The index up to which to end the check (exclusive).
      * **Returns**: `True` if the string ends with the `suffix`; `False` otherwise.
      * **Example**:
        ```python
        my_string = "Hello World"
        print(my_string.endswith("World"))
        # Output: True
        print(my_string.endswith("o World", 0, 11))
        # Output: True (checks within the specified range)
        ```

  * **`replace()`**:

      * **Purpose**: Replaces occurrences of a specified substring with a new substring. You can optionally limit the number of replacements.
      * **Syntax**: `string.replace(old_substring, new_substring, [count])`
          * `old_substring`: The substring to be replaced.
          * `new_substring`: The substring to replace `old_substring` with.
          * `count` (optional): The maximum number of times to replace `old_substring`. If omitted, all occurrences are replaced.
      * **Returns**: A new string with the replacements made.
      * **Example**:
        ```python
        my_string = "one two one three one"
        print(my_string.replace("one", "four"))
        # Output: four two four three four
        print(my_string.replace("one", "four", 1))
        # Output: four two one three one
        ```

### String Splitting Functions

These functions break a string into a list of smaller strings based on a delimiter.

  * **`split()`**:

      * **Purpose**: Splits the string into a **list of substrings** based on a specified delimiter (separator). By default, it splits by any whitespace character and discards empty strings.
      * **Syntax**: `string.split([delimiter], [maxsplit])`
          * `delimiter` (optional): The string to use as a separator. If omitted or `None`, any whitespace string (spaces, tabs, newlines) is a delimiter, and empty strings are removed from the result.
          * `maxsplit` (optional): The maximum number of splits to perform. If `maxsplit` is specified, the list will have at most `maxsplit + 1` elements.
      * **Returns**: A **list** of strings.
      * **Example**:
        ```python
        my_string = "apple banana cherry"
        print(my_string.split())
        # Output: ['apple', 'banana', 'cherry']

        my_string = "one::two::three::four"
        print(my_string.split("::", 1))
        # Output: ['one', 'two::three::four']
        ```

  * **`rsplit()`**:

      * **Purpose**: Splits the string into a list of substrings, similar to `split()`, but performs the splitting from the **right side** of the string.
      * **Syntax**: `string.rsplit([delimiter], [maxsplit])`
          * `delimiter` (optional): The string to use as a separator.
          * `maxsplit` (optional): The maximum number of splits to perform, starting from the right.
      * **Returns**: A **list** of strings.
      * **Example**:
        ```python
        my_string = "one::two::three::four"
        print(my_string.rsplit("::", 1))
        # Output: ['one::two::three', 'four']

        my_string = "python is easy to learn and also java"
        print(my_string.rsplit())
        # Output: ['python', 'is', 'easy', 'to', 'learn', 'and', 'also', 'java']
        ```
      * **Note**: Your example `a.rsplit(' ','_',2)` seems to have a typo with `'_'`. `rsplit` takes one delimiter. If you intended to split by space and then do something with an underscore, that would be a separate operation.

### String Trimming Functions

These functions are used to remove unwanted characters (typically whitespace) from the beginning, end, or both ends of a string.

  * **`strip()`**:

      * **Purpose**: Removes unwanted **leading and trailing characters** from the given string. By default, it removes leading and trailing whitespace characters (spaces, tabs, newlines, etc.).
      * **Syntax**: `string.strip([chars])`
          * `chars` (optional): A string specifying the set of characters to remove. If omitted, whitespace characters are removed.
      * **Returns**: A new string with the leading and trailing characters removed.
      * **Example**:
        ```python
        my_string = "   hello world   "
        print(my_string.strip())
        # Output: 'hello world'

        my_string = "---hello world---"
        print(my_string.strip('-'))
        # Output: 'hello world'

        my_string = "abchelloabc"
        print(my_string.strip('abc'))
        # Output: 'hello' (removes any combination of 'a', 'b', 'c' from ends)
        ```

  * **`lstrip()`**:

      * **Purpose**: Removes unwanted **leading characters** (from the left side) from the given string. By default, it removes leading whitespace characters.
      * **Syntax**: `string.lstrip([chars])`
          * `chars` (optional): A string specifying the set of characters to remove from the left.
      * **Returns**: A new string with the leading characters removed.
      * **Example**:
        ```python
        my_string = "     steve     "
        print(my_string.lstrip())
        # Output: 'steve     '

        my_string = "###data###"
        print(my_string.lstrip('#'))
        # Output: 'data###'
        ```

  * **`rstrip()`**:

      * **Purpose**: Removes unwanted **trailing characters** (from the right side) from the given string. By default, it removes trailing whitespace characters.
      * **Syntax**: `string.rstrip([chars])`
          * `chars` (optional): A string specifying the set of characters to remove from the right.
      * **Returns**: A new string with the trailing characters removed.
      * **Example**:
        ```python
        my_string = "     steve     "
        print(my_string.rstrip())
        # Output: '     steve'

        my_string = "###data###"
        print(my_string.rstrip('#'))
        # Output: '###data'
        ```

### Substring Indexing Functions

These functions are used to find the index (position) of a substring within a string.

  * **`index()`**:

      * **Purpose**: Finds the **index position of a specified substring** within the given limits in a string.
      * **Syntax**: `string.index(substring, [start], [end])`
          * `substring`: The string to search for.
          * `start` (optional): The starting index for the search.
          * `end` (optional): The ending index for the search (exclusive).
      * **Behavior**: By default, it returns the index of the *first occurrence* of the substring. **If the substring is not found within the specified limits, it will raise a `ValueError`**.
      * **Example**:
        ```python
        a = 'banana'
        print(a.index('a'))
        # Output: 1 (first 'a')

        print(a.index('a', a.index('a') + 1))
        # Output: 3 (second 'a', starting search after the first 'a')

        # To find the third 'a'
        first_a = a.index('a')
        second_a = a.index('a', first_a + 1)
        third_a = a.index('a', second_a + 1)
        print(third_a)
        # Output: 5

        # print(a.index('z')) # This would raise a ValueError
        ```

  * **`rindex()`**:

      * **Purpose**: Finds the **index position of a specified substring** within the given limits, searching from **right to left** in the string.
      * **Syntax**: `string.rindex(substring, [start], [end])`
          * `substring`: The string to search for.
          * `start` (optional): The starting index for the search.
          * `end` (optional): The ending index for the search (exclusive).
      * **Behavior**: By default, it finds the index of the *last occurrence* of the substring. **If the substring is not found within the specified limits, it will raise a `ValueError`**.
      * **Example**:
        ```python
        a = 'banana'
        print(a.rindex('a'))
        # Output: 5 (last 'a')

        # To find the second to last 'a'
        last_a = a.rindex('a')
        print(a.rindex('a', 0, last_a)) # Search from 0 up to (but not including) the last_a index
        # Output: 3
        ```

  * **`find()`**:

      * **Purpose**: Finds the **index position of a specified substring** within the given limits.
      * **Syntax**: `string.find(substring, [start], [end])`
          * `substring`: The string to search for.
          * `start` (optional): The starting index for the search.
          * `end` (optional): The ending index for the search (exclusive).
      * **Behavior**: By default, it returns the index of the *first occurrence*. **If the substring is not found in the given limits, `find()` will return `-1`**. This is a key difference from `index()`.
      * **Example**:
        ```python
        a = 'banana'
        print(a.find('a'))
        # Output: 1

        print(a.find('a', a.find('a') + 1))
        # Output: 3

        print(a.find('z'))
        # Output: -1 (no error raised)
        ```

  * **`rfind()`**:

      * **Purpose**: Finds the **index position of a given substring** from **right to left** direction of a given string.
      * **Syntax**: `string.rfind(substring, [start], [end])`
          * `substring`: The string to search for.
          * `start` (optional): The starting index for the search.
          * `end` (optional): The ending index for the search (exclusive).
      * **Behavior**: Similar to `rindex()`, but **if the substring is not found, it will return `-1`** instead of raising an error.
      * **Example**:
        ```python
        a = 'banana'
        print(a.rfind('a'))
        # Output: 5

        print(a.rfind('z'))
        # Output: -1
        ```

### String Formatting and Joining Functions

These functions are used for creating formatted strings and combining string elements.

  * **`format()`**:

      * **Purpose**: Replaces placeholders in a string with specified values. This provides a flexible way to construct strings.
      * **Syntax**: `format_string.format(arg1, arg2, ...)` or `format_string.format(**kwargs)`
          * Placeholders can be empty `{}` (positional), numbered `{0}`, `{1}` (positional by index), or named `{name}` (keyword arguments).
      * **Behavior**: The number of arguments given to a format function is typically dependent on the number of placeholders in the given format string.
      * **Returns**: A new formatted string.
      * **Example**:
        ```python
        print("My name is {} and I am {} years old.".format("Alice", 30))
        # Output: My name is Alice and I am 30 years old.

        print("My name is {0} and I am {1} years old.".format("Bob", 25))
        # Output: My name is Bob and I am 25 years old.

        print("My name is {name} and I am {age} years old.".format(name="Charlie", age=35))
        # Output: My name is Charlie and I am 35 years old.
        ```

  * **`join()`**:

      * **Purpose**: Concatenates elements of an iterable (e.g., a list or tuple of strings) into a single string, using the string on which `join()` is called as the delimiter.
      * **Syntax**: `gluestring.join(iterable)`
          * `gluestring`: The string that will be used as a separator between the elements.
          * `iterable`: An iterable containing string elements.
      * **Returns**: A new string, which is the concatenation of the elements in `iterable`.
      * **Example**:
        ```python
        words = ["Hello", "World", "Python"]
        print(" ".join(words))
        # Output: Hello World Python

        numbers = ["1", "2", "3", "4"]
        ```
